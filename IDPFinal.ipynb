```python
import numpy as np
import pandas as pd
import yfinance as yf
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt

# Define the function for loading and preparing data
def load_data(symbol, start_date="2010-01-01", end_date="2025-1-30"):
    data = yf.download(symbol, start=start_date, end=end_date)
    data = data[['Close']]
    return data

# Define a function for preprocessing
def preprocess_data(data):
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(data)
    return scaled_data, scaler

# Prepare data in sequences
def create_sequences(data, time_steps=60):
    X, y = [], []
    for i in range(time_steps, len(data)):
        X.append(data[i-time_steps:i, 0])
        y.append(data[i, 0])
    return np.array(X), np.array(y)

# Load data for top 3 NSE companies (replace 'RELIANCE.NS' with actual top 3 symbols)
symbols = ["RELIANCE.NS", "TCS.NS", "INFY.NS"]
datasets = {symbol: load_data(symbol) for symbol in symbols}

# Preprocess each dataset
preprocessed_data = {}
for symbol, data in datasets.items():
    scaled_data, scaler = preprocess_data(data)
    X, y = create_sequences(scaled_data)
    preprocessed_data[symbol] = (X, y, scaler)

# Define and compile the model
def create_model(input_shape):
    model = Sequential([
        LSTM(100, return_sequences=True, input_shape=(input_shape[1], 1)),
        Dropout(0.2),
        LSTM(50, return_sequences=False),
        Dropout(0.2),
        Dense(25),
        Dense(1)
    ])
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

# Train, evaluate, and predict for each company
results = {}
time_steps = 60
for symbol, (X, y, scaler) in preprocessed_data.items():
    # Split data into training and testing sets
    split = int(0.8 * len(X))
    X_train, X_test = X[:split], X[split:]
    y_train, y_test = y[:split], y[split:]

    # Reshape for LSTM input
    X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
    X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

    # Create and train the model
    model = create_model(X_train.shape)
    early_stop = EarlyStopping(monitor='loss', patience=10)
    model.fit(X_train, y_train, epochs=50, batch_size=32, callbacks=[early_stop])

    # Predictions
    predictions = model.predict(X_test)
    predictions = scaler.inverse_transform(predictions)
    y_test_rescaled = scaler.inverse_transform(y_test.reshape(-1, 1))

    # Calculate errors
    mse = mean_squared_error(y_test_rescaled, predictions)
    r2 = r2_score(y_test_rescaled, predictions)

    # Predict next day
    last_sequence = X_test[-1]
    next_day_pred = model.predict(np.expand_dims(last_sequence, axis=0))
    next_day_price = scaler.inverse_transform(next_day_pred)[0][0]

    # Store results and prepare for visualization
    results[symbol] = {
        "MSE": mse,
        "R2": r2,
        "Next Day Price": next_day_price,
        "Predictions": predictions,
        "Actual": y_test_rescaled
    }

    # Error Visualization
    plt.figure(figsize=(14, 5))
    plt.plot(y_test_rescaled, color='blue', label='Actual Prices')
    plt.plot(predictions, color='red', label='Predicted Prices')
    plt.title(f'{symbol} Price Prediction')
    plt.xlabel('Time')
    plt.ylabel('Price')
    plt.legend()
    plt.show()

# Display results
for symbol, metrics in results.items():
    print(f"Results for {symbol}:")
    print(f"MSE: {metrics['MSE']}")
    print(f"R2 Score: {metrics['R2']}")
    print(f"Predicted Next Day Price: {metrics['Next Day Price']}\n")
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import yfinance as yf
import pandas as pd

def create_risk_management_model():
    """
    Create a neural network for risk management scoring.
    """
    model = Sequential([
        Dense(1, input_shape=(5,), activation='linear', use_bias=False)
    ])

    # Transpose the weights to match the expected shape (5, 1)
    weights = np.array([[0.30, 0.20, 0.20, 0.15, 0.15]]).T
    model.layers[0].set_weights([weights])

    return model

def get_company_data(ticker_symbol):
    """
    Fetch financial data from Yahoo Finance.
    """
    print(f"Fetching data for {ticker_symbol}...")
    stock = yf.Ticker(ticker_symbol)

    try:
        debt_to_equity = stock.info.get('debtToEquity', 0) / 100 if stock.info.get('debtToEquity') else 0
        current_ratio = stock.info.get('currentRatio', 0) if stock.info.get('currentRatio') else 0
        operating_margin = stock.info.get('operatingMargins', 0) * 100 if stock.info.get('operatingMargins') else 0
        beta = stock.info.get('beta', 0) if stock.info.get('beta') else 0
        esg_score = stock.info.get('esgScore', 50) if stock.info.get('esgScore') else 50

        return {
            'debt_to_equity': debt_to_equity,
            'current_ratio': current_ratio,
            'operating_margin': operating_margin,
            'beta': beta,
            'esg_score': esg_score
        }
    except Exception as e:
        print(f"Error fetching data for {ticker_symbol}: {str(e)}")
        return None

def normalize_inputs(data, industry_maximums):
    """
    Normalize the input features.
    """
    return np.array([[
        1 - (data['debt_to_equity'] / industry_maximums['max_debt_to_equity']),
        data['current_ratio'] / industry_maximums['max_current_ratio'],
        data['operating_margin'] / industry_maximums['max_operating_margin'],
        1 - (data['beta'] / industry_maximums['max_beta']),
        data['esg_score'] / 100
    ]])

def main():
    # Create the model
    model = create_risk_management_model()

    # Define industry maximums
    industry_maximums = {
        'max_debt_to_equity': 2.0,
        'max_current_ratio': 2.5,
        'max_operating_margin': 20,
        'max_beta': 1.5
    }

    # Companies to analyze
    companies = {
        'Reliance': 'RELIANCE.NS',
        'TCS': 'TCS.NS',
        'Infosys': 'INFY.NS'
    }

    # Store results
    results = []

    # Analyze each company
    for company_name, ticker in companies.items():
        data = get_company_data(ticker)
        if data:
            # Normalize and predict
            normalized_inputs = normalize_inputs(data, industry_maximums)
            score = float(model.predict(normalized_inputs)[0][0])
            risk_level = "High" if score >= 0.8 else "Medium" if score >= 0.5 else "Low"

            results.append({
                'Company': company_name,
                'Ticker': ticker,
                'Risk Score': score,
                'Risk Level': risk_level,
                'Debt to Equity': data['debt_to_equity'],
                'Current Ratio': data['current_ratio'],
                'Operating Margin': data['operating_margin'],
                'Beta': data['beta'],
                'ESG Score': data['esg_score']
            })

    # Create a DataFrame with results
    df = pd.DataFrame(results)

    # Display results
    print("\nRisk Management Analysis Results:")
    print("=" * 80)
    print(df.to_string(index=False))
    print("\nDetailed Metrics:")
    print("=" * 80)
    for result in results:
        print(f"\n{result['Company']} ({result['Ticker']}):")
        print(f"Risk Score: {result['Risk Score']:.3f}")
        print(f"Risk Level: {result['Risk Level']}")
        print(f"Metrics:")
        print(f"- Debt to Equity: {result['Debt to Equity']:.2f}")
        print(f"- Current Ratio: {result['Current Ratio']:.2f}")
        print(f"- Operating Margin: {result['Operating Margin']:.2f}%")
        print(f"- Beta: {result['Beta']:.2f}")
        print(f"- ESG Score: {result['ESG Score']:.2f}")

    # Save the model
    print("\nSaving model to risk_management_model.h5...")
    model.save("risk_management_model.h5")
    print("Model saved successfully!")

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Embedding
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import yfinance as yf
import tweepy
import requests
from bs4 import BeautifulSoup
from datetime import datetime, timedelta
import re
from textblob import TextBlob

class CompanySentimentAnalyzer:
    def __init__(self):
        self.tokenizer = Tokenizer(num_words=5000, oov_token="<OOV>")
        self.max_length = 100
        self.model = self.create_sentiment_model()

        # Twitter API credentials (you need to add your own)
        self.twitter_auth = tweepy.OAuthHandler("YOUR_CONSUMER_KEY", "YOUR_CONSUMER_SECRET")
        self.twitter_auth.set_access_token("YOUR_ACCESS_TOKEN", "YOUR_ACCESS_TOKEN_SECRET")
        self.twitter_api = tweepy.API(self.twitter_auth)

    def create_sentiment_model(self):
        model = Sequential([
            Embedding(5000, 32, input_length=self.max_length),
            LSTM(64, return_sequences=True),
            LSTM(32),
            Dense(16, activation='relu'),
            Dense(1, activation='sigmoid')
        ])

        model.compile(optimizer='adam',
                     loss='binary_crossentropy',
                     metrics=['accuracy'])

        return model

    def get_news_headlines(self, company_name, days=7):
        headlines = []
        base_url = f"https://news.google.com/rss/search?q={company_name}+when:7d"

        try:
            response = requests.get(base_url)
            soup = BeautifulSoup(response.content, 'xml')
            items = soup.find_all('item')

            for item in items:
                headlines.append({
                    'title': item.title.text,
                    'date': item.pubDate.text,
                    'source': 'news'
                })
        except Exception as e:
            print(f"Error fetching news for {company_name}: {str(e)}")

        return headlines

    def get_tweets(self, company_name, count=100):
        tweets = []
        try:
            search_results = self.twitter_api.search_tweets(q=company_name,
                                                          lang="en",
                                                          count=count)

            for tweet in search_results:
                tweets.append({
                    'text': tweet.text,
                    'date': tweet.created_at,
                    'source': 'twitter'
                })
        except Exception as e:
            print(f"Error fetching tweets for {company_name}: {str(e)}")

        return tweets

    def clean_text(self, text):
        text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
        text = re.sub(r'@\w+', '', text)
        text = re.sub(r'#\w+', '', text)
        text = re.sub(r'[^\w\s]', '', text)
        text = text.lower().strip()
        return text

    def get_sentiment_score(self, text):
        analysis = TextBlob(text)
        return analysis.sentiment.polarity

    def classify_sentiment(self, sentiment_score):
        """
        Classify sentiment score into positive, negative, or neutral
        """
        if sentiment_score >= 0.1:
            return "Positive"
        elif sentiment_score <= -0.1:
            return "Negative"
        else:
            return "Neutral"

    def analyze_company_sentiment(self, company_name, ticker):
        print(f"\nAnalyzing sentiment for {company_name}...")

        news = self.get_news_headlines(company_name)
        tweets = self.get_tweets(company_name)
        all_content = news + tweets

        if not all_content:
            return None

        sentiments = []
        processed_texts = []

        for content in all_content:
            text = content.get('title', content.get('text', ''))
            cleaned_text = self.clean_text(text)

            if cleaned_text:
                sentiment = self.get_sentiment_score(cleaned_text)
                sentiments.append(sentiment)
                processed_texts.append(cleaned_text)

        avg_sentiment = np.mean(sentiments) if sentiments else 0
        sentiment_std = np.std(sentiments) if sentiments else 0

        self.tokenizer.fit_on_texts(processed_texts)
        sequences = self.tokenizer.texts_to_sequences(processed_texts)
        padded_sequences = pad_sequences(sequences, maxlen=self.max_length, padding='post')

        predictions = self.model.predict(padded_sequences)
        avg_prediction = np.mean(predictions)

        # Classify the overall sentiment
        sentiment_classification = self.classify_sentiment(avg_sentiment)

        return {
            'company': company_name,
            'ticker': ticker,
            'average_sentiment': avg_sentiment,
            'sentiment_std': sentiment_std,
            'model_prediction': avg_prediction,
            'content_count': len(all_content),
            'sentiment_classification': sentiment_classification,
            'processed_texts': processed_texts
        }

def main():
    analyzer = CompanySentimentAnalyzer()

    companies = {
        'Reliance': 'RELIANCE.NS',
        'TCS': 'TCS.NS',
        'Infosys': 'INFY.NS'
    }

    results = []

    print("\nStarting Sentiment Analysis...")
    print("=" * 80)

    for company_name, ticker in companies.items():
        result = analyzer.analyze_company_sentiment(company_name, ticker)
        if result:
            results.append(result)


            print(f"\nResults for {company_name} ({ticker}):")
            print("-" * 40)
            print(f"Sentiment Classification: {result['sentiment_classification']}")
            print(f"Average Sentiment Score: {result['average_sentiment']:.3f}")
            print(f"Content Analyzed: {result['content_count']} items")
            print(f"Confidence (Std Dev): {result['sentiment_std']:.3f}")

    #
    df = pd.DataFrame(results)

    print("\nSummary of Sentiment Analysis:")
    print("=" * 80)
    summary_df = df[['company', 'sentiment_classification', 'average_sentiment', 'content_count']]
    print(summary_df.to_string(index=False))


    print("\nSaving sentiment analysis model...")
    analyzer.model.save("sentiment_analysis_model.h5")
    print("Model saved successfully as 'sentiment_analysis_model.h5'")


    import pickle
    with open('tokenizer.pickle', 'wb') as handle:
        pickle.dump(analyzer.tokenizer, handle, protocol=pickle.HIGHEST_PROTOCOL)
    print("Tokenizer saved successfully as 'tokenizer.pickle'")

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import pandas as pd
import yfinance as yf
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt

def load_data(symbol, start_date="2010-01-01", end_date="2025-02-11"):
    data = yf.download(symbol, start=start_date, end=end_date)
    data = data[['Open', 'High', 'Low', 'Close', 'Volume']]
    return data

def preprocess_data(data):
    scaler_features = MinMaxScaler(feature_range=(0, 1))
    scaler_close = MinMaxScaler(feature_range=(0, 1))

    scaled_data = scaler_features.fit_transform(data)
    data['Close'] = scaler_close.fit_transform(data[['Close']])

    return scaled_data, scaler_features, scaler_close


def create_sequences(data, time_steps=60):
    X, y = [], []
    for i in range(time_steps, len(data)):
        X.append(data[i-time_steps:i])
        y.append(data[i, 3])
    return np.array(X), np.array(y)

symbols = ["RELIANCE.NS", "TCS.NS", "INFY.NS"]
datasets = {symbol: load_data(symbol) for symbol in symbols}

preprocessed_data = {}
for symbol, data in datasets.items():
    scaled_data, scaler_features, scaler_close = preprocess_data(data)
    X, y = create_sequences(scaled_data)
    preprocessed_data[symbol] = (X, y, scaler_features, scaler_close)

# Define the LSTM model
def create_model(input_shape):
    model = Sequential([
        LSTM(100, return_sequences=True, input_shape=(input_shape[1], input_shape[2])),
        Dropout(0.2),
        LSTM(50, return_sequences=False),
        Dropout(0.2),
        Dense(25),
        Dense(1)
    ])
    model.compile(optimizer=tf.keras.optimizers.Adam(clipvalue=1.0), loss='mean_squared_error')
    return model

# Train and predict
results = {}
time_steps = 60
for symbol, (X, y, scaler_features, scaler_close) in preprocessed_data.items():
    split = int(0.8 * len(X))
    X_train, X_test = X[:split], X[split:]
    y_train, y_test = y[:split], y[split:]

    X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], X_train.shape[2]))
    X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], X_test.shape[2]))

    model = create_model(X_train.shape)
    early_stop = EarlyStopping(monitor='loss', patience=10)
    model.fit(X_train, y_train, epochs=50, batch_size=32, callbacks=[early_stop])

    predictions = model.predict(X_test)
    predictions = scaler_close.inverse_transform(predictions)
    y_test_rescaled = scaler_close.inverse_transform(y_test.reshape(-1, 1))

    mse = mean_squared_error(y_test_rescaled, predictions)
    r2 = r2_score(y_test_rescaled, predictions)

    last_sequence = X_test[-1]
    next_day_pred = model.predict(np.expand_dims(last_sequence, axis=0))
    next_day_price = scaler_close.inverse_transform(next_day_pred)[0][0]

    results[symbol] = {
        "MSE": mse,
        "R2": r2,
        "Next Day Price": next_day_price,
        "Predictions": predictions,
        "Actual": y_test_rescaled
    }

    plt.figure(figsize=(14, 5))
    plt.plot(y_test_rescaled, color='blue', label='Actual Prices')
    plt.plot(predictions, color='red', label='Predicted Prices')
    plt.title(f'{symbol} Price Prediction')
    plt.xlabel('Time')
    plt.ylabel('Price')
    plt.legend()
    plt.show()

# Display results
for symbol, metrics in results.items():
    print(f"Results for {symbol}:")
    print(f"MSE: {metrics['MSE']}")
    print(f"R2 Score: {metrics['R2']}")
    print(f"Predicted Next Day Price: {metrics['Next Day Price']}\n")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import defaultdict

# Load the dataset
column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
iris_data = pd.read_csv('iris.data', header=None, names=column_names)
iris_data = iris_data.drop(columns=['class'])
data_points = iris_data.to_numpy()

# K-means using MapReduce-like approach
def kmeans_mapreduce(data, k, max_iterations=100):
    # Step 1: Randomly initialize centroids
    centroids = data[np.random.choice(data.shape[0], k, replace=False)]

    for _ in range(max_iterations):
        # Map Step: Assign each data point to the nearest centroid
        def map_step(point):
            distances = np.linalg.norm(centroids - point, axis=1)
            closest_centroid = np.argmin(distances)
            return closest_centroid, point

        # Apply the map step to all data points
        mapped_data = [map_step(point) for point in data]

        # Reduce Step: Compute new centroids
        def reduce_step(mapped_data):
            centroid_dict = defaultdict(list)

            # Group points by their assigned centroid
            for centroid_idx, point in mapped_data:
                centroid_dict[centroid_idx].append(point)

            # Compute new centroids as the mean of assigned points
            new_centroids = []
            for centroid_idx in sorted(centroid_dict.keys()):
                points_assigned = np.array(centroid_dict[centroid_idx])
                new_centroid = np.mean(points_assigned, axis=0)
                new_centroids.append(new_centroid)

            return np.array(new_centroids)

        # Compute new centroids
        new_centroids = reduce_step(mapped_data)

        # Check for convergence
        if np.all(centroids == new_centroids):
            break

        centroids = new_centroids

    # Assign final labels to data points
    final_labels = [map_step(point)[0] for point in data]
    return final_labels, centroids

# Run K-means
k = 3
labels, centroids = kmeans_mapreduce(data_points, k)

# Visualize the results
plt.scatter(data_points[:, 0], data_points[:, 1], c=labels, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='x', s=200, label='Centroids')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('K-means Clustering on Iris Dataset (MapReduce-like)')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import defaultdict

column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
iris_data = pd.read_csv('iris.data', header=None, names=column_names)
iris_data = iris_data.drop(columns=['class'])
data_points = iris_data.to_numpy()

def kmeans_mapreduce(data, k, max_iterations=100):
    centroids = data[np.random.choice(data.shape[0], k, replace=False)]
    plt.figure(figsize=(8, 6))

    for iteration in range(max_iterations):
        plt.clf()
        plt.scatter(data[:, 0], data[:, 1], alpha=0.5)
        plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='x', s=200, label='Centroids')
        plt.xlabel('Sepal Length')
        plt.ylabel('Sepal Width')
        plt.title(f'Iteration {iteration}')
        plt.legend()
        plt.pause(0.5)

        def map_step(point):
            distances = np.linalg.norm(centroids - point, axis=1)
            closest_centroid = np.argmin(distances)
            return closest_centroid, point

        mapped_data = [map_step(point) for point in data]

        def reduce_step(mapped_data):
            centroid_dict = defaultdict(list)
            for centroid_idx, point in mapped_data:
                centroid_dict[centroid_idx].append(point)
            new_centroids = []
            for centroid_idx in sorted(centroid_dict.keys()):
                points_assigned = np.array(centroid_dict[centroid_idx])
                new_centroid = np.mean(points_assigned, axis=0)
                new_centroids.append(new_centroid)
            return np.array(new_centroids)

        new_centroids = reduce_step(mapped_data)

        if np.all(centroids == new_centroids):
            break

        centroids = new_centroids

    final_labels = [map_step(point)[0] for point in data]
    return final_labels, centroids

k = 3
labels, centroids = kmeans_mapreduce(data_points, k)

plt.scatter(data_points[:, 0], data_points[:, 1], c=labels, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='x', s=200, label='Centroids')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Final K-means Clustering on Iris Dataset')
plt.legend()
plt.show()
```
